```python
text_list = list(elocution_df['text'])
#text_list
```

```python
def find_ngrams(input_list, n):
    return list(zip(*[input_list[i:] for i in range(n)]))

n_gram_list = []

for book in text_list:
    split_book_list = book.split(' ')
    n = 4
    input_list = split_book_list
    output_list = find_ngrams(input_list, n)
    n_gram_list.append(output_list)
```

```python
# making a new column
elocution_df['ngrams'] = n_gram_list
```

```python
elocution_df[0:10]
```

```python
corpus_ngrams = []

for book in n_gram_list:
    corpus_ngrams.extend(book)
   
        
```

```python
# simpler version learning version 2


text = elocution_df['text']
model = CountVectorizer(ngram_range = (4, 4))
matrix = model.fit_transform(text).toarray()
df_output = pd.DataFrame(data = matrix, columns = model.get_feature_names_out())
df_output.T.tail(5)
```

```python
df_output["title"]=elocution_df["title"]
```

```python
df_output=df_output.set_index("title")
```

```python
df_output=df_output.T
```

```python
def floor(num):
    if num>0:
        return 1
    else:
        return 0
```

```python
df_output=df_output.applymap(floor)
```

```python
df_output["sum"]=df_output.sum(axis=1)
```

```python
df_output["sum"]
```

```python
df_output=df_output[df_output["sum"]>2].sort_values("sum", ascending=False)
```
